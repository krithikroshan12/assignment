```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
url="https://www.flipkart.com/search?q=shoes&as=on&as-show=on&otracker=AS_Query_TrendingAutoSuggest_2_0_na_na_na&otracker1=AS_Query_TrendingAutoSuggest_2_0_na_na_na&as-pos=2&as-type=TRENDING&suggestionId=shoes&requestId=d42d8e58-1650-40ff-8e74-730a29ae934c"
```

```python
permission=requests.get(url)
```

```python
permission
```

```python
link=BeautifulSoup(permission.text)
```

```python
link
```

```python
product=link.select("._2WkVRV")
```

```python
product
```

```python
price=link.select("._3bPFwb ._30jeq3")
```

```python
price
```

```python
import pandas as pd
```

```python
productlist=pd.DataFrame(product,columns=["name"])
```

```python
productlist
```

```python
productlist["rate"]=price
```

```python
productlist
```

```python
productlist.to_csv("C:/krithik roshan/assignment.csv")
```
